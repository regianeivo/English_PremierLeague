<h2> Data Collection </h2>
Data from the English Premier League (2010 to 2023), collected from https://www.football-data.co.uk/englandm.php </br>
I concanated all the files that are set per year utilizing pandas library (pd) to one dataframe called 'matches' </br>
The size of the dataframe is 4676 rows and 139 columns </br>

```python
#importing necessary python libraries to explore the dataset

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#data collection for 12 years of English Premier League

file1 = pd.read_csv('PL_2010_2011.csv', header = 0)
file2 = pd.read_csv('PL_2011_2012.csv', header = 0)
file3 = pd.read_csv('PL_2012_2013.csv', header = 0)
file4 = pd.read_csv('PL_2013_2014.csv', header = 0)
file5 = pd.read_csv('PL_2014_2015.csv', header = 0)
file6 = pd.read_csv('PL_2015_2016.csv', header = 0)
file7 = pd.read_csv('PL_2016_2017.csv', header = 0)
file8 = pd.read_csv('PL_2017_2018.csv', header = 0)
file9 = pd.read_csv('PL_2018_2019.csv', header = 0)
file10 = pd.read_csv('PL_2019_2020.csv', header = 0)
file11 = pd.read_csv('PL_2020_2021.csv', header = 0)
file12 = pd.read_csv('PL_2021_2022.csv', header = 0)
file13 = pd.read_csv('PL_2022_2023.csv', header = 0)
```

```python
#creating a dataframe concatenating data from all files collected above

files = [file1, file2, file3, file4, file5, file6, file7, file8, file9, file10, file11, file12, file13]
matches = pd.concat(files)
matches.shape
```
